## First Dataset for EVs
This dataset shows the make, model year, EV type, range, and retail price

```python
# Showing columns 5-8 & 10-11
ev_df = pd.read_csv("Data/ev_pop.csv")[["Model Year", "Make", "Electric Vehicle Type", "Electric Range", "Base MSRP"]]
ev_df.head(90)
```

```python
# Dropping all rows that are plug-in hybrids
def no_plug_ins(df):
    plug_ins = df.loc[df["Electric Vehicle Type"] == "Plug-in Hybrid Electric Vehicle (PHEV)"].index
    [plug_ins.astype(int) for x in plug_ins]
    df.drop(index=plug_ins, inplace=True)
    
no_plug_ins(ev_df)
```
